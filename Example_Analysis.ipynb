## 2. Import Libraries

We'll use the following Python libraries for our analysis:

```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Statistical analysis
from scipy import stats

# Configuration
plt.style.use('seaborn-v0_8')
%matplotlib inline
```

## 3. Load and Explore Data

Let's start by loading our dataset and performing initial exploration.

```python
# Load data (example)
# df = pd.read_csv('sales_data.csv')

# For demonstration, create sample data
df = pd.DataFrame({
    'Date': pd.date_range('2023-01-01', periods=100, freq='D'),
    'Product': np.random.choice(['A', 'B', 'C'], 100),
    'Sales': np.random.randint(100, 1000, 100),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 100)
})

# Display first few rows
df.head()
```

### Data Summary

Let's examine the basic statistics of our dataset:

```python
# Dataset information
print("Dataset Shape:", df.shape)
print("\nColumn Types:")
print(df.dtypes)
print("\nStatistical Summary:")
df.describe()
```

## 4. Data Cleaning

### Check for Missing Values

Before analysis, we need to ensure data quality:

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values per Column:")
print(missing_values)

# Calculate percentage
missing_percentage = (missing_values / len(df)) * 100
print("\nMissing Percentage:")
print(missing_percentage)
```

> **Note:** No missing values found in the dataset. We can proceed with the analysis.

## 5. Exploratory Data Analysis

### 5.1 Sales by Product

Let's analyze which products generate the most revenue:

```python
# Group by product and calculate total sales
product_sales = df.groupby('Product')['Sales'].agg(['sum', 'mean', 'count'])
product_sales.columns = ['Total Sales', 'Average Sale', 'Transactions']
product_sales = product_sales.sort_values('Total Sales', ascending=False)
print(product_sales)
```

### 5.2 Regional Analysis

Understanding sales distribution across regions:

```python
# Sales by region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("Total Sales by Region:")
print(region_sales)
```

## 6. Visualization

### 6.1 Sales Trend Over Time

```python
# Plot sales over time
plt.figure(figsize=(12, 6))
df.set_index('Date')['Sales'].plot()
plt.title('Sales Trend Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### 6.2 Product Performance Comparison

```python
# Bar chart of product sales
plt.figure(figsize=(10, 6))
product_sales['Total Sales'].plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Total Sales by Product', fontsize=16, fontweight='bold')
plt.xlabel('Product')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Statistical Analysis

### Sales Distribution

Let's examine if sales follow a normal distribution:

**Null Hypothesis (Hâ‚€):** Sales data follows a normal distribution  
**Alternative Hypothesis (Hâ‚):** Sales data does not follow a normal distribution  
**Significance Level:** Î± = 0.05

```python
# Perform normality test
statistic, p_value = stats.normaltest(df['Sales'])
print(f"Normality Test Results:")
print(f"  Test Statistic: {statistic:.4f}")
print(f"  P-value: {p_value:.4f}")

if p_value > 0.05:
    print("  âœ“ Result: Fail to reject Hâ‚€ (Data appears normally distributed)")
else:
    print("  âœ— Result: Reject Hâ‚€ (Data does not appear normally distributed)")
```
